```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
model.train(
    data="final_crop/yolo_datasets/classes.yaml",
    epochs=300,
    patience=20,
    device=0,
    project="YOLOv8",
    name="YOLOv8_crop_default",
    verbose=True
)
```

```python
model.val("final_crop/yolo_datasets/classes.yaml", split="test")
```
