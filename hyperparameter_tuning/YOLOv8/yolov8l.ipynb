```python
from ultralytics import YOLO
```

```python
# class weights
cls_weights = [
    1.71051282, 1.71051282, 1.71051282, 1.69918492, 0.60980849,
    0.20493364, 1.71051282, 1.70766671, 1.69637635, 1.62647812,
    1.71051282, 1.71051282, 1.68247163
]
```

```python
# Hyperparameters
params = [
    {'batch': 8, 'lr': 0.001, 'optimizer': 'Adam'},
    {'batch': 8, 'lr': 0.001, 'optimizer': 'SGD'},
    {'batch': 8, 'lr': 0.01, 'optimizer': 'Adam'},
    {'batch': 8, 'lr': 0.01, 'optimizer': 'SGD'},
    {'batch': 16, 'lr': 0.001, 'optimizer': 'Adam'},
    {'batch': 16, 'lr': 0.001, 'optimizer': 'SGD'},
    {'batch': 16, 'lr': 0.01, 'optimizer': 'Adam'},
    {'batch': 16, 'lr': 0.01, 'optimizer': 'SGD'}
]
```

Trial 1

```python
# trial 1
i = 0

model = YOLO("yolov8l.pt")

model.train(
    data="yolo_datasets/classes.yaml", 
    epochs=150, 
    patience=150, 
    cls_weights=cls_weights,
    pretrained=True,
    batch=params[i]["batch"], 
    lr0=params[i]["lr"], 
    optimizer=params[i]["optimizer"], 
    device=0,
    project="YOLOv8",
    name=f"YOLOv8_{i}",
    verbose=True
)
```

```python
# trial 1
model.val(split="test")
```
